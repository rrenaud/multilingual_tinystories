```python
spanish_tinystories_ds = datasets.load_dataset("robrenaud/multilingual_tinystories")
```

```python
with open('train.txt', 'w') as full_stories_output :
    for story in tqdm.tqdm(spanish_tinystories_ds['train']['story']):
        full_stories_output.write(story)
```

```python
tokenizer = tokenizers.ByteLevelBPETokenizer()

tokenizer.train(files=['train.txt'], vocab_size=2**13, min_frequency=2)

tokenizer.save_model('.', 'tiny-stories-spanish-bpe')
```
